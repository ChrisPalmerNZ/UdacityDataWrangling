So for the last row in that table:

- p1_dog is whether or not the #1 prediction is a breed of dog → TRUE

````python
def getScore(stringtweet):
    """Returns the numerator and denominator of the last fraction found in the passed string."""
    
    stringtweet = stringtweet.encode('ascii','ignore').decode('utf-8').rstrip().lstrip()
    stringtweet = re.sub(r"http\S+", "", stringtweet).rstrip().lstrip()
    stringtweet = stringtweet.replace("\n"," ").replace(","," ").replace(";", " ")\
                             .replace("(", " ").replace(")", " ").replace("10s", "10 s")
    
    stringtweetrev = stringtweet[::-1]
    findstring = re.findall(r'[0-9]+/[0-9]+', stringtweetrev)

    if len(findstring) == 0:
        stringscoreNumerator = 0
        stringscoreDenominator = 0    
    else:    
        stringscore  = findstring[0][::-1] 
        strPosn = stringtweet.find(stringscore)
       
        if strPosn > 0 and \
            stringtweet[strPosn-1] == '.' and \
            stringtweet[strPosn-2].isdigit():
            stringscore = "." + stringscore
            strPosn = strPosn - 1
            while strPosn > 0 and stringtweet[strPosn-1].isdigit():
                stringscore = stringtweet[strPosn-1] + stringscore
                strPosn = strPosn - 1
                
        stringscoreSplit = stringscore.split("/")
        stringscoreNumerator = float(stringscoreSplit[0])
        stringscoreDenominator = int(stringscoreSplit[1])
        
    return stringscoreNumerator, stringscoreDenominator        
````

### Process the downloaded twitter json
- Get tweet_id, text, retweet and favourite information, and the correct dog scores
- Created two versions of this, one uses `json.loads`, the other `pd.read_json` - in order to understand how to work with them

````python
def json_to_list(jsonfile, showProgress=True):
    """Processes passed file using json.loads and returns id, full_text, retweet_count, 
       favorite_count, urls, retweet data and dog_score in a list."""
    tweets_list = []
    with open(jsonfile, 'r', encoding='UTF-8') as f:
        for line in f:
            jsdata = json.loads(line) 
            recid = jsdata["id"]
            rtcount = jsdata["retweet_count"]
            fvcount = jsdata["favorite_count"]
            twttext = jsdata["full_text"]
            twt_urls = ', '.join(re.findall(r"http\S+", twttext)) #+ ' '
            if showProgress:
                print(recid, "...")
                
            twt_expandedurls = None
            
            # get image from media if possible    
            if "extended_entities" in line and "media" in line and "expanded_url" in line:
                if "retweeted_status" in line and "quoted_status" in line:
                    twt_expandedurls = jsdata["retweeted_status"]["quoted_status"]["extended_entities"]["media"][0]["expanded_url"]
                elif "retweeted_status" in line:
                    twt_expandedurls = jsdata["retweeted_status"]["extended_entities"]["media"][0]["expanded_url"]
                elif "quoted_status" in line:
                    twt_expandedurls = jsdata["quoted_status"]["extended_entities"]["media"][0]["expanded_url"]
                else:    
                    twt_expandedurls = jsdata["extended_entities"]["media"][0]["expanded_url"]
             
            # just get url from entites if no image file is found in media
            if twt_expandedurls is None and ("entities" in line and "urls" in line and "expanded_url" in line):
                urls = dict(*jsdata["entities"]["urls"])
                twt_expandedurls = urls.get("expanded_url") 
                
            dogscoreNum, dogscoreDen = getScore(twttext)

            if "retweeted_status" in line:
                retweeted_status = jsdata["retweeted_status"]
                rt_id = int(retweeted_status["id"])
                rt_userid = retweeted_status["user"]["id_str"]
                rt_timestamp = retweeted_status["created_at"]
            else:
                rt_id = None
                rt_userid = None
                rt_timestamp = None
                
            if showProgress:
                print(recid, rtcount, fvcount, dogscoreNum, dogscoreDen, twttext, twt_urls, 
                      twt_expandedurls, rt_id, rt_userid, rt_timestamp)
                
            tweets_list.append([recid, rtcount, fvcount, dogscoreNum, dogscoreDen, twttext, twt_urls, 
                                twt_expandedurls, rt_id, rt_userid, rt_timestamp])
            
        return tweets_list
````

````python
def json_df_to_list(jsonfile, showProgress=True):
    """Processes passed file as a converted dataframe using pd.read_json and returns id, 
       full_text, retweet_count, favorite_count, urls, retweet data and dog_score in a list."""
    tweets_list = []
    with open(jsonfile, 'r', encoding='UTF-8') as f:
        tweets_json = pd.read_json(f, lines=True)
        for jsdata in tweets_json.itertuples():
            recid = jsdata.id
            rtcount = jsdata.retweet_count
            fvcount = jsdata.favorite_count
            twttext = jsdata.full_text
            twt_urls = ', '.join(re.findall(r"http\S+", twttext)) #+ ' '
            if showProgress:
                print(recid, "...")

            twt_expandedurls = None
            
            # get image from media if possible
            if pd.notnull(jsdata.extended_entities):
                extended_dict = dict(jsdata.extended_entities)
                extended_media = dict(extended_dict.get("media")[0])
                if extended_media is not None:
                    twt_expandedurls = extended_media.get('expanded_url')
                    
            if twt_expandedurls is None and pd.notnull(jsdata.retweeted_status):
                retweeted_status = dict(jsdata.retweeted_status)
                extended_dict = retweeted_status.get("extended_entities")
                if extended_dict is None:
                    quoted_status = retweeted_status.get("quoted_status")
                    if quoted_status is not None:
                        extended_dict = quoted_status.get("extended_entities")
                        
                if extended_dict is not None:
                    extended_dict = dict(extended_dict)
                    extended_media = dict(extended_dict.get("media")[0])
                    if extended_media is not None:
                        twt_expandedurls = extended_media.get('expanded_url')
                        
            if twt_expandedurls is None and pd.notnull(jsdata.quoted_status):
                quoted_status = dict(jsdata.quoted_status)
                if quoted_status is not None:
                    extended_dict = quoted_status.get("extended_entities")
                    if extended_dict is not None:
                        extended_dict = dict(extended_dict)
                        extended_media = dict(extended_dict.get("media")[0])
                        if extended_media is not None:
                            twt_expandedurls = extended_media.get('expanded_url')
                    
            # just get url from entites if no image file is found in media
            if twt_expandedurls is None and pd.notnull(jsdata.entities):
                ent_dict = dict(jsdata.entities)
                urls_dict = dict(*ent_dict.get("urls"))
                twt_expandedurls = urls_dict.get("expanded_url") 
                
            dogscoreNum, dogscoreDen = getScore(twttext)

            rt_id = None
            rt_userid = None
            rt_timestamp = None
                
            if pd.notnull(jsdata.retweeted_status):
                retweeted_status = jsdata.retweeted_status
                rt_id = int(retweeted_status["id"])
                rt_userid = retweeted_status["user"]["id_str"]
                rt_timestamp = retweeted_status["created_at"]
                    
            if showProgress:
                print(recid, rtcount, fvcount, dogscoreNum, dogscoreDen, twttext, twt_urls, 
                      twt_expandedurls, rt_id, rt_userid, rt_timestamp)
                
            tweets_list.append([recid, rtcount, fvcount, dogscoreNum, dogscoreDen, twttext, twt_urls, 
                                twt_expandedurls, rt_id, rt_userid, rt_timestamp])
            
        return tweets_list
````
